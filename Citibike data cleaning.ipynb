```python
import pandas as pd
import datetime as dt
```

```python
csv_path = "Resources/202209-citibike-tripdata.csv"
bike = pd.read_csv(csv_path)
bike.head()
```

```python
#bike.dtypes
```

```python
bike['started_at']=bike['started_at'].astype("datetime64")
bike['ended_at']=bike['ended_at'].astype("datetime64")
```

```python
#bike.dtypes
```

```python
tripduration=(bike['ended_at']-bike['started_at']).dt

tripduration=tripduration.seconds

tripduration.head()
```

```python
bike.insert(loc=4,
          column='trip_duration',
          value=tripduration)
bike.head()
```

```python
bike1 = bike[(bike['started_at'] > "2022-09-27")]
bike1
```

```python
bike1.to_csv('Resources/202209-citibike.csv',index=False)
```

```python
bike3= bike[(bike['started_at'] > "2022-09-24")]
bike3.sort_values('started_at')
```

```python
bike3.to_csv('Resources/202209-7daysCitibike.csv',index=False)
```
